#  02 Timestamp Alignment

This module will allow you to align timestamps with FP data. In order to run this module, you must first process photometry traces using the 01.pipeline_process module. 

The following timestamp formats are currently supported. Please refer to the Synaptech Suite Guide for more information on what file types are required for each timestamp method, and what the format of those files should be.

1. Keydown Timestamps
2. Timestamps generated by Arduino using a serial string read node in Bonsai
3. Video Frame numbers for post-hoc hand scored behaviors

This pipeline will generate a single file with timestamps aligned to FP data. Please use module 03.pipeline_pooling to segment data around timestamps and generate peri-event plots.

## Initialize Timestamp Alignment

```python
from routine.oo_interface import NPMAlign
align = NPMAlign()
```

## Select processed data file 
Run cell to enable file upload. Choose the processed data file generated by the 01.pipeline_process module.

```python
align.set_data()
```

## Specify where to save timestamps aligned with FP data

```python
align.set_paths()
```

## Select Timestamp File(s) to align. 
Run cell to enable file selection. Use the file browser to navigate your cloud workspace. Click the blue downward arrow button to open the selected folder. CTRL + click to select multiple files, then click the white double arrow button to add to selected files.

For more information on what files to include, please refer to the Photometry Analysis Pipeline Guide. Please note that some timestamp formats require multiple files to be uploaded. Please select all necessary files.

```python
align.set_ts()
```

## Align timestamps with FP system clock

```python
align.align_data()
```

## Write out aligned timestamp file

```python
align.export_data()
```
